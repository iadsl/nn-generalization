```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical

# Load CIFAR-10 data
(x_train, y_train), (x_test, y_test) = cifar10.load_data()


# Create the ResNet model
model = ResNet50(input_shape=(32, 32, 3), classes=10, weights=None)

# Compile the model with SGD optimizer
model.compile(optimizer=SGD(learning_rate=0.01, momentum=0.9),
              loss=SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=50, batch_size=128, validation_data=(x_test, y_test))
```

```python
import tensorflow as tf
from tensorflow.keras.optimizers import Optimizer

class SAM(Optimizer):
    def __init__(self, optimizer, rho=0.05, **kwargs):
        super(SAM, self).__init__(**kwargs)
        self.optimizer = optimizer
        self.rho = rho

    def get_gradients(self, loss, params):
        grads = self.optimizer.get_gradients(loss, params)
        e_w = [(tf.stop_gradient(w) + self.rho * tf.stop_gradient(g)) for w, g in zip(params, grads)]
        loss_e_w = loss(e_w)
        return [(g + g_e_w) / 2 for g, g_e_w in zip(grads, self.optimizer.get_gradients(loss_e_w, params))]

    def apply_gradients(self, grads_and_vars, **kwargs):
        self.optimizer.apply_gradients(grads_and_vars, **kwargs)

    def get_config(self):
        config = {'rho': self.rho}
        base_config = super(SAM, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))

# Integrate SAM with ResNet
sam_optimizer = SAM(SGD(learning_rate=0.01, momentum=0.9), name="sam")

model.compile(optimizer=sam_optimizer,
              loss=SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=50, batch_size=128, validation_data=(x_test, y_test))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Function to compute loss values around the minimum
def compute_loss_surface(model, x, y, param1_range, param2_range, direction1, direction2):
    loss_surface = np.zeros((len(param1_range), len(param2_range)))
    for i, p1 in enumerate(param1_range):
        for j, p2 in enumerate(param2_range):
            new_weights = [w + p1*d1 + p2*d2 for w, d1, d2 in zip(model.trainable_weights, direction1, direction2)]
            model.set_weights(new_weights)
            loss_surface[i, j] = model.evaluate(x, y, verbose=0)[0]
    return loss_surface

# Directions for perturbing weights
direction1 = [np.random.randn(*w.shape) for w in model.trainable_weights]
direction2 = [np.random.randn(*w.shape) for w in model.trainable_weights]

param_range = np.linspace(-0.1, 0.1, 21)

# Compute loss surfaces
loss_surface_sgd = compute_loss_surface(model, x_test, y_test, param_range, param_range, direction1, direction2)
loss_surface_sam = compute_loss_surface(model, x_test, y_test, param_range, param_range, direction1, direction2)

# Plot the loss surfaces
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

X, Y = np.meshgrid(param_range, param_range)

ax1.plot_surface(X, Y, loss_surface_sgd, cmap='viridis')
ax1.set_title('SGD - Sharp Minimum')

ax2.plot_surface(X, Y, loss_surface_sam, cmap='viridis')
ax2.set_title('SAM - Wide Minimum')

plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.datasets import cifar10
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load CIFAR-10 data
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Define the ResNet model
def create_resnet_model():
    return ResNet50(input_shape=(32, 32, 3), classes=10, weights=None)

# Train model with SGD
sgd_model = create_resnet_model()
sgd_model.compile(optimizer=SGD(learning_rate=0.01, momentum=0.9),
                  loss=SparseCategoricalCrossentropy(from_logits=True),
                  metrics=['accuracy'])
sgd_model.fit(x_train, y_train, epochs=1, batch_size=128, validation_data=(x_test, y_test))

# Train model with SAM (Custom SAM implementation)
class SAM(tf.keras.optimizers.Optimizer):
    def __init__(self, optimizer, rho=0.05, **kwargs):
        super(SAM, self).__init__(**kwargs)
        self.optimizer = optimizer
        self.rho = rho

    def get_gradients(self, loss, params):
        grads = self.optimizer.get_gradients(loss, params)
        e_w = [(tf.stop_gradient(w) + self.rho * tf.stop_gradient(g)) for w, g in zip(params, grads)]
        loss_e_w = loss(e_w)
        return [(g + g_e_w) / 2 for g, g_e_w in zip(grads, self.optimizer.get_gradients(loss_e_w, params))]

    def apply_gradients(self, grads_and_vars, **kwargs):
        self.optimizer.apply_gradients(grads_and_vars, **kwargs)

    def get_config(self):
        config = {'rho': self.rho}
        base_config = super(SAM, self).get_config()
        return dict(list(base_config.items()) + list(config.items()))

sam_optimizer = SAM(SGD(learning_rate=0.01, momentum=0.9),name="sam")

sam_model = create_resnet_model()
sam_model.compile(optimizer=sam_optimizer,
                  loss=SparseCategoricalCrossentropy(from_logits=True),
                  metrics=['accuracy'])
sam_model.fit(x_train, y_train, epochs=1, batch_size=128, validation_data=(x_test, y_test))

# Function to compute loss values around the minimum
def compute_loss_surface(model, x, y, param1_range, param2_range, direction1, direction2):
    loss_surface = np.zeros((len(param1_range), len(param2_range)))
    for i, p1 in enumerate(param1_range):
        for j, p2 in enumerate(param2_range):
            new_weights = [w + p1*d1 + p2*d2 for w, d1, d2 in zip(model.trainable_weights, direction1, direction2)]
            model.set_weights(new_weights)
            loss_surface[i, j] = model.evaluate(x, y, verbose=0)[0]
    return loss_surface

# Generate random directions for perturbing weights
direction1 = [np.random.randn(*w.shape) for w in sgd_model.trainable_weights]
direction2 = [np.random.randn(*w.shape) for w in sgd_model.trainable_weights]

param_range = np.linspace(-0.1, 0.1, 21)

# Compute loss surfaces for both models
loss_surface_sgd = compute_loss_surface(sgd_model, x_test, y_test, param_range, param_range, direction1, direction2)
loss_surface_sam = compute_loss_surface(sam_model, x_test, y_test, param_range, param_range, direction1, direction2)

# Plot the loss surfaces
fig = plt.figure(figsize=(12, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

X, Y = np.meshgrid(param_range, param_range)

ax1.plot_surface(X, Y, loss_surface_sgd, cmap='viridis')
ax1.set_title('SGD - Sharp Minimum')

ax2.plot_surface(X, Y, loss_surface_sam, cmap='viridis')
ax2.set_title('SAM - Wide Minimum')

plt.show()
```
